```python
from matplotlib import pyplot as plt

list_time = [58.3, 60.9, 303, 609, 2478]
hash_time = [54.9, 57.3, 65.8, 69.7, 84.3]
tst_time = [82.7, 92.7, 160, 1510, 4420]

list_time = [x/5 for x in list_time]
hash_time = [x/5 for x in hash_time]
tst_time = [x/5 for x in tst_time]

data_size = ["10", "500", "5k", "50k", "200k"]

plt.plot(data_size, list_time, "r-")
plt.plot(data_size, hash_time, "b-")
plt.plot(data_size, tst_time, "g-")
plt.xlabel('Number of elements')
plt.ylabel('Time per operation (ms)')
plt.title('Insert word-frequency')
plt.show()
```
